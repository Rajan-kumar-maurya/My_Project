```python
import matplotlib.pyplot as plt

# Create a figure with 2x2 subplots
fig, axs = plt.subplots(2, 2)

# Show the figure
plt.show()
```

```python
# Data for the bar chart
categories = ['Category A', 'Category B', 'Category C', 'Category D', 'Category E']
values = [23, 56, 41, 62, 19]

# Create a figure with 2x2 subplots
fig, axs = plt.subplots(2, 2)

# Plot the bar chart in the first subplot
axs[0, 0].bar(categories, values)

# Set title and labels for the subplot
axs[0, 0].set_title('Bar Chart')
axs[0, 0].set_xlabel('Categories')
axs[0, 0].set_ylabel('Values')

# Adjust spacing between subplots
fig.tight_layout()

# Show the figure
plt.show()
```

```python
# Data for the histogram
data = [12, 17, 21, 18, 14, 13, 16, 9, 12, 15, 19, 11, 14, 16, 20, 18, 15, 13, 16, 11, 10]

# Create a figure with 2x2 subplots
fig, axs = plt.subplots(2, 2)

# Plot the histogram in the second subplot
axs[0, 1].hist(data)

# Set title and labels for the subplot
axs[0, 1].set_title('Histogram')
axs[0, 1].set_xlabel('Data')
axs[0, 1].set_ylabel('Frequency')

# Adjust spacing between subplots
fig.tight_layout()

# Show the figure
plt.show()
```

```python
# Data for the pie chart
labels = ['Apple', 'Banana', 'Orange', 'Mango']
sizes = [30, 25, 15, 30]

# Create a figure with 2x2 subplots
fig, axs = plt.subplots(2, 2)

# Plot the pie chart in the third subplot
axs[1, 0].pie(sizes, labels=labels, autopct='%1.1f%%')

# Set title for the subplot
axs[1, 0].set_title('Pie Chart')

# Adjust spacing between subplots
fig.tight_layout()

# Show the figure
plt.show()
```

```python
# Data for the scatter plot
x_values = [1, 2, 3, 4, 5]
y_values = [2, 5, 3, 6, 4]

# Create a figure with 2x2 subplots
fig, axs = plt.subplots(2, 2)

# Plot the scatter plot in the fourth subplot
axs[1, 1].scatter(x_values, y_values)

# Set title and labels for the subplot
axs[1, 1].set_title('Scatter Plot')
axs[1, 1].set_xlabel('X values')
axs[1, 1].set_ylabel('Y values')

# Adjust spacing between subplots
fig.tight_layout()

# Show the figure
plt.show()

```

```python
import matplotlib.pyplot as plt

# Data for the plots
categories = ['Category A', 'Category B', 'Category C', 'Category D', 'Category E']
values = [23, 56, 41, 62, 19]
hist_data = [12, 17, 21, 18, 14, 13, 16, 9, 12, 15, 19, 11, 14, 16, 20, 18, 15, 13, 16, 11, 10]
pie_labels = ['Apple', 'Banana', 'Orange', 'Mango']
pie_sizes = [30, 25, 15, 30]
scatter_x = [1, 2, 3, 4, 5]
scatter_y = [2, 5, 3, 6, 4]

# Create a figure with 2x2 subplots
fig, axs = plt.subplots(2, 2)

# Plot the bar chart in the first subplot
axs[0, 0].bar(categories, values)
axs[0, 0].set_title('Bar Chart')
axs[0, 0].set_xlabel('Categories')
axs[0, 0].set_ylabel('Values')

# Plot the histogram in the second subplot
axs[0, 1].hist(hist_data, bins=5)
axs[0, 1].set_title('Histogram')
axs[0, 1].set_xlabel('Data')
axs[0, 1].set_ylabel('Frequency')

# Plot the pie chart in the third subplot
axs[1, 0].pie(pie_sizes, labels=pie_labels, autopct='%1.1f%%')
axs[1, 0].set_title('Pie Chart')

# Plot the scatter plot in the fourth subplot
axs[1, 1].scatter(scatter_x, scatter_y)
axs[1, 1].set_title('Scatter Plot')
axs[1, 1].set_xlabel('X values')
axs[1, 1].set_ylabel('Y values')

# Adjust spacing between subplots
plt.subplots_adjust(wspace=0.4, hspace=0.6)

# Show the figure
plt.show()
```
