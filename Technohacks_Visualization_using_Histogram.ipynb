```python
column_names = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'class']
df = pd.read_csv('Iris.csv', header=None, names=column_names)
```

```python
# Plot histogram for each feature
plt.figure(figsize=(10, 6))
plt.hist(df['sepal_length'], bins=10, alpha=0.5, label='Sepal Length')
plt.hist(df['sepal_width'], bins=10, alpha=0.5, label='Sepal Width')
plt.hist(df['petal_length'], bins=10, alpha=0.5, label='Petal Length')
plt.hist(df['petal_width'], bins=10, alpha=0.5, label='Petal Width')
plt.xlabel('Measurement')
plt.ylabel('Frequency')
plt.title('Histogram of Iris Dataset Features')
plt.legend()
plt.show()
```
